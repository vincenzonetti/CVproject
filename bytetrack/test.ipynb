```python
import cv2
from ultralytics import YOLO

model = YOLO("yolo11s.pt") 

video_path = "data\\videos\\out13.mp4"
cap = cv2.VideoCapture(video_path)

# Make the display window resizable and reasonably sized
cv2.namedWindow("YOLO11s Detection", cv2.WINDOW_NORMAL)
cv2.resizeWindow("YOLO11s Detection", 960, 540)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    print(frame.shape)

    results = model.track(frame, persist=True, line_width=1, tracker="bytetrack.yaml")
    annotated_frame = results[0].plot()

    
    cv2.imshow("YOLO11s Detection", annotated_frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows() 
```
